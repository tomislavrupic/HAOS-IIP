# Geometry Emergence Probe

This notebook visualizes the deterministic output of the geometry emergence sweep.

```python
from pathlib import Path
import json

import matplotlib.pyplot as plt
import numpy as np

RESULTS_PATH = Path("../diagnostics/geometry_probe_results.json")
payload = json.loads(RESULTS_PATH.read_text(encoding="utf-8"))
results = payload["results"]

kernel_widths = np.array([item["kernel_width"] for item in results], dtype=float)
distortion = np.array([item["distortion_score"] for item in results], dtype=float)
effective_dimension = np.array([item["effective_dimension"] for item in results], dtype=float)
retention_matrix = np.array([item["step_retention"] for item in results], dtype=float)
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(kernel_widths, distortion, marker="o", linewidth=2)
ax.set_title("Distortion vs Kernel Width")
ax.set_xlabel("Kernel width")
ax.set_ylabel("Distortion score")
ax.grid(alpha=0.3)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(kernel_widths, effective_dimension, marker="s", linewidth=2, color="#b85c38")
ax.set_title("Effective Dimension Regimes")
ax.set_xlabel("Kernel width")
ax.set_ylabel("Effective dimension")
ax.grid(alpha=0.3)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(9, 4.5))
image = ax.imshow(retention_matrix, aspect="auto", origin="lower", cmap="viridis")
ax.set_title("Neighborhood Stability Heatmap")
ax.set_xlabel("Transport step transition")
ax.set_ylabel("Kernel width index")
ax.set_yticks(np.arange(len(kernel_widths)))
ax.set_yticklabels([f"{value:.2f}" for value in kernel_widths])
fig.colorbar(image, ax=ax, label="Retention")
plt.tight_layout()
```
